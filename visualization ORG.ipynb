```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your TMDb 10,000 dataset
# Make sure to adjust the path to where your dataset is located
df = pd.read_csv('10000 Movies Data')
```

```python
# Ensure the 'release_date' column is in datetime format
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
# Extract the year from the 'release_date'
df['year'] = df['release_date'].dt.year
```

```python
# Drop rows where the 'year' is NaN (which could happen for invalid release dates)
df = df.dropna(subset=['year'])
```

```python
# Plot the number of movies per year
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='year', hue='year', palette='coolwarm', legend=False)
```

```python
# Add labels and title
plt.title('Movies by Release Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()
```

```python
# Assuming the 'genres' column contains a string with genres separated by commas like "Action, Comedy, Drama"
df['Genres'] = df['Genres'].fillna('')  # Handle any missing genres

# Split the genres into individual genres, then flatten the list and count occurrences
genres_series = df['Genres'].str.split(',', expand=True).stack()

# Count the frequency of each genre
genre_counts = genres_series.value_counts()

# Plotting - Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='magma')

# Add labels and title
plt.title('Top Genres by Frequency (Bar Chart)', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()

# Show the bar chart
plt.show()

# Plotting - Pie Chart (Top 10 genres)
plt.figure(figsize=(8, 8))
genre_counts.head(10).plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('magma', n_colors=10), 
                           wedgeprops={'edgecolor': 'black'})

# Add title
plt.title('Top 10 Genres by Frequency (Pie Chart)', fontsize=16)
plt.ylabel('')  # Remove the y-axis label

# Show the pie chart
plt.tight_layout()
plt.show()
```

```python
# Ensure the 'vote_average' column is numeric and handle any missing values
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')

# Drop rows where the 'vote_average' is NaN
df = df.dropna(subset=['vote_average'])

# Plotting the histogram for rating distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['vote_average'], bins=20, kde=True, color='blue', stat='density', palette='magma')

# Add labels and title
plt.title('Rating Distribution of Movies', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Ensure that 'budget' and 'revenue' are numeric and handle any missing values
df['Budget'] = pd.to_numeric(df['Budget'], errors='coerce')
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')

# Drop rows where either 'budget' or 'revenue' is NaN (invalid data)
df = df.dropna(subset=['Budget', 'Revenue'])

# Plotting the scatter plot for Budget vs. Revenue
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Budget', y='Revenue', data=df, color='purple', alpha=0.6)

# Add labels and title
plt.title('Budget vs. Revenue', fontsize=16)
plt.xlabel('Budget (in millions)', fontsize=12)
plt.ylabel('Revenue (in millions)', fontsize=12)

# Log scale for better visualization of the spread (optional, especially if there are extreme values)
plt.xscale('log')
plt.yscale('log')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Ensure that 'revenue' is numeric and handle any missing values
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')

# Drop rows where 'revenue' is NaN (invalid data)
df = df.dropna(subset=['Revenue'])

# Plotting the Revenue Distribution (Histogram)
plt.figure(figsize=(12, 6))

# Histogram for Revenue Distribution
sns.histplot(df['Revenue'], bins=30, kde=True, color='green', stat='density', palette='mako')

# Add labels and title
plt.title('Revenue Distribution of Movies', fontsize=16)
plt.xlabel('Revenue (in millions)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)

# Show the histogram plot
plt.tight_layout()
plt.show()

# Plotting the Revenue Distribution (Box Plot)
plt.figure(figsize=(12, 6))

# Box Plot for Revenue Distribution
sns.boxplot(x=df['Revenue'], color='lightblue')

# Add labels and title
plt.title('Revenue Distribution (Box Plot)', fontsize=16)
plt.xlabel('Revenue (in millions)', fontsize=12)

# Show the box plot
plt.tight_layout()
plt.show()
```

```python
# Ensure that 'budget' is numeric and handle any missing values
df['Budget'] = pd.to_numeric(df['Budget'], errors='coerce')

# Drop rows where 'budget' is NaN (invalid data)
df = df.dropna(subset=['Budget'])

# Plotting the Budget Distribution (Histogram)
plt.figure(figsize=(12, 6))

# Histogram for Budget Distribution
sns.histplot(df['Budget'], bins=30, kde=True, color='orange', stat='density', palette='magma')

# Add labels and title
plt.title('Budget Distribution of Movies', fontsize=16)
plt.xlabel('Budget (in millions)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)

# Show the histogram plot
plt.tight_layout()
plt.show()

# Plotting the Budget Distribution (Box Plot)
plt.figure(figsize=(12, 6))

# Box Plot for Budget Distribution
sns.boxplot(x=df['Budget'], color='lightgreen')

# Add labels and title
plt.title('Budget Distribution (Box Plot)', fontsize=16)
plt.xlabel('Budget (in millions)', fontsize=12)

# Show the box plot
plt.tight_layout()
plt.show()
```

```python
# Ensure the 'release_date' column is in datetime format and the 'vote_average' column is numeric
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')

# Extract the year from the 'release_date' column
df['year'] = df['release_date'].dt.year

# Drop rows where 'vote_average' or 'year' are NaN
df = df.dropna(subset=['vote_average', 'year'])

# Group by 'year' and calculate the average rating for each year
average_rating_by_year = df.groupby('year')['vote_average'].mean()

# Plotting the line graph for Average Rating by Year
plt.figure(figsize=(12, 6))
sns.lineplot(x=average_rating_by_year.index, y=average_rating_by_year.values, marker='o', color='blue')

# Add labels and title
plt.title('Average Rating by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
